# Extract Buttons, modifiers etc


```python

#print(parsed_profile['keyDiffs'])
import functools
import re

def check_base_input(input: dict):

    if 'added' not in input.keys():
        return None
        
    name = input['name']
    key, modifiers = derive_key(list(input['added'].values()))
    return name,key,modifiers

def derive_key(input: dict):
    _logger.warning("derive_key: {}".format(input))
    key = re.findall(r'\d+',input[0]['key'])

    _logger.warning(f"Key is: {key}")
    modifier = set(input[0]['reformers'].values()) if 'reformers' in input[0].keys() else None
    return key, modifier
```

# Add buttons/modifers to device

```python
ipt.clear_devices()
ipt.add_device('03F2F260-B49D-11ea-8001-444553540000', 'test')

# import pprint
# pp = pprint.PrettyPrinter(indent=4)
# pp.pprint(parsed_profile)

for x in parsed_profile['keyDiffs'].values():
    
    y = check_base_input(x)

    if y is not None:

        if y[2] is None:
            _logger.warning(f'No modifier detected in {y[2]} for {y[1]}')
            ipt.add_inputs('03F2F260-B49D-11ea-8001-444553540000', input_id=y[1], style=ipt.InputTypes.BUTTON, command=y[0], modifiers=[])
        else:
            ipt.add_input_modifier('03F2F260-B49D-11ea-8001-444553540000', input_id=y[1], command=y[0], modifier=(y[2]))

device = ipt._get_device('03F2F260-B49D-11ea-8001-444553540000')

inputs = device.get_device_inputs()

for x in inputs:
    print(x.identifier)
    print(x.command)
    print(x.modifiers)

```
